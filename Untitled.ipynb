```python
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats, optimize, interpolate
from scipy.integrate import odeint
from scipy.integrate import solve_bvp
plt.rcParams.update({'font.size': 18})
nz=1000
L = 1
h = L/(nz+1)
Zrange = np.linspace(0,1,nz)
rows, cols = (nz, nz)
arr = [[0 for i in range(cols)] for j in range(rows)]   # Initializing the system matrix A
A = np.array(arr)                                       
for i in range(rows):
    for j in range(cols):
        if j==i:
            A[i][j] = 2
            if not i==0:
                # print(A1zi)
                A[i][j-1] = -1
            if not j==cols-1:
                A[i][j+1] = -1
            else:
                A[i][j] = 2 - 1                 # boundary condition: C(n+1)=C(n) ==> A3(z_n)C(n+1) + A2(z_n)C_n + A1(z_n)C(n-1) = 0 --> (A3(z_n)+A2(z_n))C_n + A1(z_n)C(n-1) = 0

print(A)

f = np.ones(nz)*h**2
C_0 = 5
f[0] += C_0
C = np.linalg.solve(A, f)
# print(C)
Cs = C_0 - 0.5*Zrange**2 + L*Zrange

plt.plot(Zrange,C)
plt.plot(Zrange,Cs,color="red")
```
