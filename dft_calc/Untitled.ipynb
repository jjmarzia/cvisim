```python
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt
from numpy import loadtxt

lower_x = np.linspace(0,5.72777778e-01,2063)
mean_x = np.linspace(0,2.75888889e+01,99321)
upper_x = np.linspace(0,1.31388889e+03,47386)


lower_y = loadtxt("dft_lower.txt", comments="#", delimiter=",", unpack=False)
mean_y = loadtxt("dft_mean.txt", comments="#", delimiter=",", unpack=False)
upper_y = loadtxt("dft_upper.txt", comments="#", delimiter=",", unpack=False)
```

```python
fig, ax = plt.subplots(figsize = (5.8,2.7), layout='constrained')

ax.plot(lower_x, lower_y*1e6,'--',color='red',label=r'$0.493$eV lower bound')
ax.plot(mean_x, mean_y*1e6,color='red',label=r'$0.918$eV mean')
ax.plot(upper_x, upper_y*1e6,'-.',color='red',label=r'$1.342$eV upper bound')

plt.xlabel('log t hrs')
plt.ylabel(r'$r_p|_{z=0} \ \mu$m')
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
# plt.ylim(0,0.2)
ax.set_xscale("log")
# ax2.set_xscale("log")
# ax.set_yscale("log")
# ax2.set_yscale("log")
plt.xticks([0.57,27.59, 1316.11],['0.57','27.59', '1316.11'])
ax.grid(True)
plt.savefig('FigWalker.pdf', dpi=220,bbox_inches='tight')
plt.savefig('FigWalker.svg', dpi=220,bbox_inches='tight')
```
