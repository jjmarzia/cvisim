```python
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt



a1_05 = np.loadtxt("alpha1-0.5.txt", dtype=float)
a1_05 = a1_05.ravel()
a1_1 = np.loadtxt("alpha1-1.txt", dtype=float)
a1_1 = a1_1.ravel()
a1_2 = np.loadtxt("alpha1-2.txt", dtype=float)
a1_2 = a1_2.ravel()


a10_05 = np.loadtxt("alpha10-0.5.txt", dtype=float)
a10_05 = a10_05.ravel()
a10_1 = np.loadtxt("alpha10-1.txt", dtype=float)
a10_1 = a10_1.ravel()
a10_2 = np.loadtxt("alpha10-2.txt", dtype=float)
a10_2 = a10_2.ravel()

L = 5e-3
zspace = np.linspace(0,L,len(a1_05))
```

```python
fig, (ax, ax2) = plt.subplots(2, 1, sharex=True, figsize = (5,2.7), layout='constrained')

plt.scatter(zspace, a1_05,label='dt 0.5s')
plt.scatter(zspace, a1_1,label='dt 1s')
plt.scatter(zspace, a1_2,label='dt 2s')

ax.set_ylim(.80, 1.)  # outliers only
ax2.set_ylim(0, .25)
ax.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax.xaxis.tick_top()
ax.tick_params(labeltop=False)  # don't put tick labels at the top
ax2.xaxis.tick_bottom()
d = .015
kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
ax.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal
kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal
plt.xlabel('z m')
plt.ylabel(r'$\xi$')
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
ax.set_title(r'$\xi_0 = 0.30, \ \alpha=1$') 
# ax.grid(True)
# ax2.grid(True)
plt.savefig('dtall_alpha1.svg', dpi=220,bbox_inches='tight')
plt.savefig('dtall_alpha1.pdf', dpi=220,bbox_inches='tight')
plt.show()
```

```python
fig, (ax, ax2) = plt.subplots(2, 1, sharex=True, figsize = (5,2.7), layout='constrained')

plt.scatter(zspace, a10_05,label='dt 0.5s')
plt.scatter(zspace, a10_1,label='dt 1s')
plt.scatter(zspace, a10_2,label='dt 2s')

ax.set_ylim(.80, 1.)  # outliers only
ax2.set_ylim(0, .25)
ax.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax.xaxis.tick_top()
ax.tick_params(labeltop=False)  # don't put tick labels at the top
ax2.xaxis.tick_bottom()
d = .015
kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
ax.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal
kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal
plt.xlabel('z m')
plt.ylabel(r'$\xi$')
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
ax.set_title(r'$\xi_0 = 0.30, \ \alpha=10$') 
# ax.grid(True)
# ax2.grid(True)
plt.savefig('dtall_alpha10.svg', dpi=220,bbox_inches='tight')
plt.savefig('dtall_alpha10.pdf', dpi=220,bbox_inches='tight')
plt.show()
```

```python
norm_05_2_alpha1 = np.linalg.norm(a1_05-a1_2)
norm_05_1_alpha1 = np.linalg.norm(a1_05-a1_1)

norm_05_2_alpha10 = np.linalg.norm(a10_05-a10_2)
norm_05_1_alpha10 = np.linalg.norm(a10_05-a10_1)

norms_a1 = np.array([norm_05_2_alpha1, norm_05_1_alpha1])
norms_a10 = np.array([norm_05_2_alpha10, norm_05_1_alpha10])

finemesh = np.array([2, 1])

fig, ax = plt.subplots(figsize = (5.8,2.7), layout='constrained')

ax.scatter(finemesh, norms_a1,color='red',label=r'$\alpha=1$')
ax.scatter(finemesh, norms_a10,color='blue',label=r'$\alpha=10$')

plt.xlabel('dt s')
plt.ylabel(r'$||\xi_{dt} - \xi_{0.5}||$')
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
# plt.ylim(0,0.2)
ax.set_xscale("log")
# ax2.set_xscale("log")
ax.set_yscale("log")
# ax2.set_yscale("log")
plt.xticks([1,2],['1','2'])
plt.xlim([0.7,2.9])
plt.ylim([6e-3, 3e-1])
ax.grid(True)
plt.savefig('dt_loglog.pdf', dpi=220,bbox_inches='tight')
plt.savefig('dt_loglog.svg', dpi=220,bbox_inches='tight')
```
