```python
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt
import scipy

sig_y = np.array([93, 129, 145, 158, 189, 233])*1e6
d = np.array([416, 106, 75.6, 43.2, 29.8, 16.0])*1e-6
d05 = d**-0.5

plt.plot(d05,sig_y,label='data')
plt.plot(np.unique(d05), np.poly1d(np.polyfit(d05, sig_y, 1))(np.unique(d05)),label='best fit line')
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
```
