```python
fig, ax = plt.subplots(figsize = (4,2), layout='constrained')

norm28_224 = np.linalg.norm(cy-by)
norm112_224 = np.linalg.norm(ay-by)
print('norm28=',norm28_224,' norm112=',norm112_224)
normsy = np.array([norm28_224, norm112_224])
normsx = np.array([28, 112])
plt.scatter(1/normsx, 1/normsy,color='red')

plt.xlabel('element spacing')
plt.ylabel(r'$h/||\sigma_{xx,h} - \sigma_{xx,224}||$')
# plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
# plt.ylim(0,0.2)
ax.set_xscale("log")
# ax2.set_xscale("log")
ax.set_yscale("log")
# ax2.set_yscale("log")
plt.xticks([1/28,1/112],['1/28','1/112'])
plt.xlim([1/300,1/10])
plt.ylim([1e-8, 1e-5])
ax.grid(True)
plt.savefig('h_loglog.pdf', dpi=220,bbox_inches='tight')
plt.savefig('h_loglog.svg', dpi=220,bbox_inches='tight')
```
